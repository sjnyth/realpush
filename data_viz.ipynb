```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

# barplot and countplot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips,palette='Set1')
```

```python
sns.boxplot(x='day',y='total_bill', data=tips,palette='coolwarm', hue='smoker')
```

```python
sns.boxplot(x='day',y='tip', data=tips,palette='coolwarm', hue='smoker')
```

```python
sns.boxplot(data=tips,palette='Set3', orient='h')
```

```python
tips['percent']=tips['tip']/tips['total_bill']
```

```python
tips.head(1)
```

```python
sns.boxplot(x='day',y='percent', data=tips,palette='coolwarm', hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill', data=tips,palette='coolwarm', hue='sex')
```

# violinplot

```python
sns.violinplot(x='day',y='total_bill', data=tips, palatte='coolwarm',hue='sex')
```

```python
sns.violinplot(x='day',y='percent', data=tips, palatte='coolwarm',hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, palatte='rainbow',hue='sex', split=True)
```

# strip plot

stripplot will draw a scatter plot where one var is cater

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips, hue='sex',palette='Set1',split=True)
```

```python

sns.violinplot(x='tip',y='day',data=tips)
sns.swarmplot(x='tip',y='day',data=tips,color='white',size=4)
```

```python

sns.violinplot(x='day',y='tip',data=tips)
sns.swarmplot(x='day',y='tip',data=tips,color='white',size=4)
```
