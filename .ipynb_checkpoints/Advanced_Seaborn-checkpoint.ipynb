```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris_df=sns.load_dataset('iris')
```

```python
iris_df.head(1)
```

```python
#scatterpot x='sepal_width', y='sepal_length'
sns.scatterplot(x='sepal_width',y='sepal_length', data=iris_df, hue='species')
```

```python
palette_dict={x:'gray' for x in iris_df['species'].unique()}
```

```python
palette_dict
```

```python
palette_dict['virginica']='red'
```

```python
palette_dict
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='sepal_width',y='sepal_length', data=iris_df, hue='species', palette=palette_dict, size='petal_length')
```
